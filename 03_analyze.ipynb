```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from brain.v3.brain import Brain
```

```python
pre_trained = Brain(learning_rate=0)
pre_trained.load_model(name='pre-trained/all-learnable/pre-trained')
```

```python
pre_trained_learned = Brain(learning_rate=0)
pre_trained_learned.load_model(name='pre-trained/all-learnable')
```

```python
from ml_tools.keras.plot import plot_conv2d

cfg = {
 "vmin": -0.2,
 "vmax": 0.2
}

_, [ax1, ax2] = plt.subplots(ncols=2, figsize=(15,5))

ax1.set_title("pre_trained")
plot_conv2d(pre_trained.model.layers[1], ax=ax1, **cfg)

ax2.set_title("pre_trained_learned")
plot_conv2d(pre_trained_learned.model.layers[1], ax=ax2, **cfg)

plt.show()
```
